```python
def convert_sqft_to_num(x):
    tokens=x.split('-')
    if len(tokens)==2:
            return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
                return None
            
```

```python
def remove_pps_outliers(df):
    df_out=pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m= np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df=subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df6=remove_pps_outliers(df5)
df6.shape
```

```python
def plot_scatter_chart(df,location):
    bhk2=df[(df.location==location) & (df.bhk==2)]
    bhk3=df[(df.location==location) & (df.bhk==3)]
    bhk4=df[(df.location==location) & (df.bhk==4)]
    bhk5=df[(df.location==location) & (df.bhk==5)]
    plt.rcParams['figure.figsize']=(10,8)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHk',s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,color='green',marker='*',label='3 BHk',s=50)
    plt.scatter(bhk4.total_sqft,bhk4.price,color='red',marker='+',label='4 BHk',s=50)
    plt.scatter(bhk5.total_sqft,bhk5.price,color='black',marker='o',label='5 BHk',s=50)
    plt.xlabel('Total square feet area')
    plt.ylabel('Price (Lakh Indian Rupees)')
    plt.title(location)
    plt.legend()
```

```python
def remove_bhk_outliers(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
              }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df7 = remove_bhk_outliers(df6)
df7.shape
```

```python
df9 = df8.drop(['size','price_per_sqft'],axis=1)
df9.head()
```

```python
dummies = pd.get_dummies(df9.location)
dummies.head()
```

```python
df10 = pd.concat([df9, dummies.drop('other',axis=1)],axis=1)
df10.head()
```

```python
df11 = df10.drop('location',axis=1)
df11.head()
```

```python
x = df11.drop(['price'],axis=1)
x.head()
```

```python
y = df11[['price']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
LinearReg = LinearRegression()
LinearReg.fit(xtrain,ytrain)
LinearReg.score(xtest,ytest)
```

```python
from sklearn.model_selection import ShuffleSplit, cross_val_score

CV = ShuffleSplit(n_splits=5,test_size=0.2,random_state=10)

a = cross_val_score(LinearRegression(),x,y,cv=CV)
np.average(a)
```

```python
def predict_price(location,sqft,bath,bhk):
    loc_index = np.where(x.columns==location)[0][0]
    
    new = np.zeros(len(x.columns))
    new[0] = sqft
    new[1] = bath
    new[2] = bhk
    
    if loc_index>=0:
        new[loc_index] = 1
    
    return LinearReg.predict([new])[0]
```
